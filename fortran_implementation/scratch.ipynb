```python
import jax

jax.config.update("jax_enable_x64", True)
import jax.numpy as jnp
import matplotlib.pyplot as plt
```

```python
with open("fortran_implementation/times.txt", "w") as f:
    for i in jnp.linspace(-1,1,100):
        f.write(str(float(i)) + "\n")
```

```python
from squishyplanet import OblateSystem

state = {
    "times" : jnp.linspace(-1,1,100),
    "a" : 200.0,
    "e" : 0.3,
    "i" : 89.75 * jnp.pi / 180,
    "Omega" : 95 * jnp.pi / 180,
    "omega" : jnp.pi / 3.5,
    "period" : 1001.0,
    "t0" : 0.2,
    "parameterize_with_projected_ellipse" : True,
    "projected_effective_r" : 0.1,
    "projected_f" : 0.3,
    "projected_theta" : 0.2,
    "tidally_locked" : False,
    "ld_u_coeffs" : jnp.array([0.008, 0.007, 0.006, 0.005, 0.004, 0.003, 0.002, 0.001])
}

p = OblateSystem(**state)

g = p.lightcurve()

plt.plot(g)
```
